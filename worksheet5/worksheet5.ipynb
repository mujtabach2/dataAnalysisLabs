```python

import seaborn as sns
import matplotlib.pyplot as plt
d_diam = sns.load_dataset('diamonds')
d_ice = sns.load_dataset('seaice')

```

With d_diam, use Seaborn to create a scatterplot with carat (x), price(y), cut (hue).

Title the plot "The price of diamonds", and label the x and y axes "Weight (carats)" and "Price ($)", respectively. Your figure should like the image below:

```python
sns.scatterplot(x='carat', y='price', data=d_diam, hue='cut')
plt.title('The price of diamonds')
plt.xlabel('Weight (carats)')
plt.ylabel('Price ($)')
plt.show()
```

With d_ice, use Matplotlib to create a line plot using the Extent column, with a dashed black line

add x-ticks at 0, 3000, 6000, 9000, 12000. Give the ticks labels: "1980", "1992", "2000", "2008", "2016"

Give the x and y-axes labels: "Year", "Thickness (m)" respectively. Give the plot the title "Change in Artic sea ice thickness over time". Your figure should like the image below:

```python

plt.plot(d_ice['Extent'], linestyle='--', color='black')
ticks = [0, 3000, 6000, 9000, 12000]
labels = ["1980", "1992", "2000", "2008", "2016"]
plt.xticks(ticks, labels)

plt.xlabel('Year')
plt.ylabel('Thickness (m))')
plt.title("Change in Artic sea ice thickness over time")

plt.show()
```
